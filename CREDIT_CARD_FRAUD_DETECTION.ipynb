```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading dataset to dataframe
credit_card_csv = pd.read_csv("creditcard.csv")
```

```python
#getting first five row data
credit_card_csv.head()
```

```python
#getting last five row data
credit_card_csv.tail()
```

```python
#dataset information
credit_card_csv.info()
```

```python
#checking for null values
credit_card_csv.isnull().sum()
```

```python
#dropping NA values
credit_card = credit_card_csv.dropna()
```

```python
#checking after dropping for null values
credit_card.isnull().sum()
```

```python
#Checking for distribution of lefit amd fradulent transactions
credit_card['Class'].value_counts()
```

This Dataset is highly unbalanced as
0.0 --> legit transactions
1.0 --> fradulent transactions

```python
#seperating data fro analysis
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for both transactions
credit_card.groupby('Class').mean()
```

Building a sample dataset containing similar distribuiton of lefit and fradulent transactions...

no of fradulent transactions = 142 therefore taking equal number of legit transactions for balanced dataset creation..

```python
legit_sample = legit.sample(n=142)
```

```python
#merging two datasets
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean
```

```python
#Seperating dataset into training and testing dataset
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
#training data over LogisticRegression
model.fit(x_train, y_train)
```

```python
#predicting score on training data
x_train_prediction = model.predict(x_train)
training_data_acc = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on training data :', training_data_acc)
```

```python
#predicting score on testing data
x_test_prediction = model.predict(x_test)
test_data_acc = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on testing data :', test_data_acc)
```
